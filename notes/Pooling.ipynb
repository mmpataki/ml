### Pooling implementations

We here try different types of pooling techniques average-pooling, max-pooling, min-pooling. These pooling algorithms are used in the convolution neural networks. The motto of this work is to visualize the pooling (compaction of image information)

```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
from scipy import ndimage

#suppress deprecated warnings from the scipy ndimage.
import warnings
warnings.filterwarnings('ignore')
```

#### Read the image which will be used further

```python
img = ndimage.imread('poolingsample.jpeg')
ih, iw, d = img.shape
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.imshow(img)
plt.show()
```

#### Hyperparameters.
We are going to use these hyperparameters.

```python
stride = 2
winsize = 2
```

#### A generic pooling function.

We define a generic pooling function where we supply the function we want to use as a parameter.

```python
def pool(pfunc):
    nh, nw = m.floor(ih / winsize), m.floor(iw / winsize)
    apimg = np.zeros([nh, nw, 3])
    for i in range(nh):
        y = i * winsize
        for j in range(nw):
            x = j * winsize
            apimg[i][j][0] = pfunc(img[y:(y+winsize), x:(x+winsize), 0])
            apimg[i][j][1] = pfunc(img[y:(y+winsize), x:(x+winsize), 1])
            apimg[i][j][2] = pfunc(img[y:(y+winsize), x:(x+winsize), 2])
    return apimg
```

#### Average pooling
Now we use average pooling to get a smaller version of the above image.

```python
apimg = pool(np.average)
plt.imshow(apimg)
plt.show()
```

#### Maxpooling

For each window we choose maximum of the window here

```python
apimg = pool(np.max)
plt.imshow(apimg)
plt.show()
```

As you can see the maxpooling saves some of image information while average pooling isn't so good. Well may be we can find a better scenario where average pooling works better

#### Min-pooling

We will choose minimum of the window here

```python
apimg = pool(np.min)
plt.imshow(apimg)
plt.show()
```
